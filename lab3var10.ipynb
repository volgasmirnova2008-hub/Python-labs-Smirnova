```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

```python
sunspots = sm.datasets.sunspots.load_pandas()
s_data = sunspots.data
s_data
```

```python
data = s_data[(s_data.YEAR > 1989) & (s_data.YEAR < 2009)]
data
```

```python

plt.plot(data.YEAR, data.SUNACTIVITY, marker ='o', color = 'gold', linestyle = '-', linewidth = 2, markersize = 5)
plt.title('Солнечная активность')
plt.xlabel('Год')
plt.ylabel('Солнечная активность')
ax = plt.gca()
ax.xaxis.set_major_locator(MultipleLocator(2)) 
ax.xaxis.set_minor_locator(MultipleLocator(1))   
ax.yaxis.set_minor_locator(MultipleLocator(10))
ax.grid(True, which='major', linestyle='-')    
ax.grid(True, which='minor', linestyle='--') 
plt.tight_layout()
plt.show()
```

```python
plt.bar(data.YEAR, data.SUNACTIVITY, color = 'gold')
plt.title('Солнечная активность')
plt.xlabel('Год')
plt.ylabel('Солнечная активность')
ax = plt.gca()
ax.xaxis.set_major_locator(MultipleLocator(2)) 
ax.xaxis.set_minor_locator(MultipleLocator(1))   
ax.yaxis.set_minor_locator(MultipleLocator(10))
ax.grid(True, which='major', linestyle='-')  
```

```python
plt.hist(data.SUNACTIVITY, color = 'gold')
plt.title("Солнечная активность")
plt.xlabel("Значения активности")
plt.ylabel("Частота лет")
ax = plt.gca()
ax.xaxis.set_major_locator(MultipleLocator(20)) 
ax.xaxis.set_minor_locator(MultipleLocator(10))  
```
